```python
#remove all rows that contain atleast one missing value
df_cleaned = df.dropna()
```

```python
#check if there are any missing values still left

df.isnull().sum()
```

```python
#LotFrontage has 259 missing values. Since lot sizes have outliers, using a median rather than mean is a better option.
df["LotFrontage"].fillna(df["LotFrontage"].median(), inplace=True)

```

```python
#check for missing values again
df.isnull().sum()
```

```python
#now that we dont have any missing values, lets visualize the data 

import matplotlib.pyplot as plt
import seaborn as sns

#we are creating a histogram that shows how values of LotFrontage are distributed 

plt.figure(figsize=(10, 5))
sns.histplot(df["LotFrontage"], bins=30, kde=True, color="blue")
plt.title("Distribution of LotFrontage")
plt.xlabel("Lot Frontage")
plt.ylabel("Count")
plt.show()

#the kernel density curve 
```

```python
#boxplots are the ideal graph type to spot outliers

plt.figure(figsize=(8, 5))
sns.boxplot(x=df["LotFrontage"], color="blue")
plt.title("Boxplot of LotFrontage")
plt.show()

```

```python
#we are using a scatterplot to show the relationship between LotFrontage and SalePrice

plt.figure(figsize=(10, 6))
sns.scatterplot(x=df["LotFrontage"], y=df["SalePrice"], color="red")
plt.title("LotFrontage vs SalePrice")
plt.xlabel("Lot Frontage")
plt.ylabel("Sale Price")
plt.show()

```

```python
#we are trying to find how many properties are in each neighborhood

plt.figure(figsize=(12, 5))
sns.countplot(y=df["Neighborhood"], order=df["Neighborhood"].value_counts().index, palette="viridis")
plt.title("Distribution of Properties by Neighborhood")
plt.show() 
```

```python
#here we try to explore if sale price and lot area are correlated
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.scatterplot(x=df["LotArea"], y=df["SalePrice"], color="green", alpha=0.5)
plt.title("Sale Price vs. Lot Area")
plt.xlabel("Lot Area")
plt.ylabel("Sale Price")
plt.show()
```

```python
#here we try to explore the relationship between year built and sale price
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.scatterplot(x=df["YearBuilt"], y=df["SalePrice"], color="blue", alpha=0.5)
plt.title("Sale Price vs. Year Built")
plt.xlabel("Year Built")
plt.ylabel("Sale Price")
plt.show() 
```

```python
#group year built into nearest decade
df["DecadeBuilt"] = (df["YearBuilt"] // 10) * 10 

#plot graph of sale prices in each decade

plt.figure(figsize=(12, 6))
sns.boxplot(x=df["DecadeBuilt"], y=df["SalePrice"], palette="viridis")
plt.xticks(rotation=45)
plt.title("Sale Price by Decade Built")
plt.xlabel("Decade Built")
plt.ylabel("Sale Price")
plt.show()
```

```python
#we are creating a historgram showing the average sale price of homes in each neighborhood

avg_price_by_neighborhood = df.groupby("Neighborhood")["SalePrice"].mean().sort_values()

plt.figure(figsize=(14, 6))
sns.barplot(x=avg_price_by_neighborhood.index, y=avg_price_by_neighborhood.values, palette="coolwarm")
plt.xticks(rotation=90)
plt.title("Average Sale Price by Neighborhood")
plt.xlabel("Neighborhood")
plt.ylabel("Average Sale Price")
plt.show()
```

```python
#we are visualizing the relationship between the neighborhood and lot area.
#this will help provide insight on whether certain neighborhoods have bigger lot area sizes and hence this sale price.

avg_lotarea_by_neighborhood = df.groupby("Neighborhood")["LotArea"].mean().sort_values()

plt.figure(figsize=(14, 6))
sns.barplot(x=avg_lotarea_by_neighborhood.index, y=avg_lotarea_by_neighborhood.values, palette="coolwarm")
plt.xticks(rotation=90)
plt.title("Average Lot Area by Neighborhood")
plt.xlabel("Neighborhood")
plt.ylabel("Average Lot Area (sq ft)")
plt.show()

```

```python
#we are visualizing the relationship between neighborhood and lotfrontage
#this will provide insight on whether certain neighborhoods have larger lot frontages and hence this affects sale price of homes.

avg_lotfrontage_by_neighborhood = df.groupby("Neighborhood")["LotFrontage"].mean().sort_values()

plt.figure(figsize=(14, 6))
sns.barplot(x=avg_lotfrontage_by_neighborhood.index, y=avg_lotfrontage_by_neighborhood.values, palette="coolwarm")
plt.xticks(rotation=90)
plt.title("Average Lot Frontage by Neighborhood")
plt.xlabel("Neighborhood")
plt.ylabel("Average Lot Frontage (feet)")
plt.show()
```
